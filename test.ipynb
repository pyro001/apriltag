# Project : get the best fit for torque
## define data

 input data torque (originally kg force)

output data gear ratio 1:y

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([0.016,0.130,6.8, 23.000 ])

y = np.array([1, 16, 256, 4096])


```

## convert  input data to Nm (torque)
the arm is 20 cm in length 
since the input was measured in kg force, we convert it to newtons where 1 kg-force = 9.80665 N

```python
x= (9.80665*x)*0.2 
```

# find line of best fit
a line is of the form y = mx+c

where m is the slope of the line

c is the y axis intercept

np.polyfit fits a first degree polynomial (a line) to the x,y data, yeilding the slope(m) and the y- intercept (c)

```python

m, c = np.polyfit(x, y, 1)


```

We then plot the points on a scatter plot along the generated line

```python
# add points to plot
plt.scatter(x, y,label= "observation")
plt.xlabel("torque")
plt.ylabel("gear ratio")
plt.title("estimation of best fit for motor data")

# add line of best fit to plot
plt.plot(x, m*x+c,label= "fit line")
plt.legend(loc="best", fontsize='7')
plt.show()
```

now given that we have a best fit available, we predict the value at  gear ratio 1:1 which would be the natural stall torque of the motor. 

To do this, we use y= mx+c ;
where m,c was acquired with the polynomial fit
and y=1 and x is unknown

which would change the Equation to : 
 x= (y-c)/m
or better still, x= (1-c)/m

```python
x_estimated= (1-c)/m
print("the estimated stall torque is: ", x_estimated)
```
